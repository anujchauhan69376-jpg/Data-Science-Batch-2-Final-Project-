```python
# Loading DataSet
import pandas as pd
import numpy as np
df = pd.read_csv("loan.csv", low_memory=False)

```

```python
# Exploring The dataset
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
# Check Missing Values
print(df.isnull().sum())
```

```python
#Required Columns
columns = [
    "loan_amnt",
    "funded_amnt",
    "term",
    "int_rate",
    "grade",
    "sub_grade",
    "annual_inc",
    "purpose",
    "loan_status",
    "dti",
    "installment",
    "home_ownership",
    "verification_status",
    "emp_length",
    "addr_state"
]

df = df[columns]
```

```python
print(df.shape)
```

```python
# Checking duplicate rows
print("Duplicate rows:", df.duplicated().sum())
```

```python
#Keep Only Completed Loans
df = df[df["loan_status"].isin(["Fully Paid", "Charged Off"])]
```

```python
print(df["loan_status"].value_counts())
print(df.shape)
```

```python
print(df.isnull().sum())
```

```python
df["emp_length"] = df["emp_length"].fillna("Unknown")
```

```python
df = df.dropna(subset=[
    "annual_inc",
    "loan_amnt",
    "funded_amnt",
    "int_rate",
    "dti",
    "installment"
])
```

```python
#Remove Invalid Values
df = df[df["annual_inc"] > 0]

df = df[df["loan_amnt"] > 0]

df = df[df["funded_amnt"] > 0]

df = df[df["int_rate"] > 0]

df = df[df["dti"] >= 0]

df = df[df["installment"] > 0]
```

```python
#Creating Income Band
bins = [0, 50000, 100000, float("inf")]

labels = [
    "Low Income",
    "Medium Income",
    "High Income"
]

df["Income Band"] = pd.cut(
    df["annual_inc"],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
print(df["Income Band"].value_counts())
```

```python
df["Default"] = (df["loan_status"] == "Charged Off").astype(int)
```

```python
print(df["Default"].value_counts())
```

```python
#Taking only 30k rows
df = df.sample(n=30000, random_state=42, replace=True)

df.reset_index(drop=True, inplace=True)
```

```python
print(df.shape)
```

```python
print(df.isnull().sum())
```

```python
print(df["loan_status"].value_counts())
```
